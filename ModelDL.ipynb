# Download and use GPT2

This notebook is to download the GPT-2 model as a PKL file and reload it to make sure it works. This was done in a Jupyter Notebook file to take advantage of Google Colab's GPUs. See https://huggingface.co/gpt2


### Step 1: Install and Import Dependencies

```python
#imports
from transformers import GPT2Tokenizer, GPT2Model, GPT2LMHeadModel
from google.colab import files
import pickle 
import torch
```

### Step 2: Download and Save Model

```python
#import model and dump it as a pickle file
#this saves the file in the notebook environment
model = GPT2LMHeadModel.from_pretrained('gpt2', return_dict=True)
pickle.dump(model, open('model.pkl', 'wb'))
```

```python
#download the pickle file to local environment
files.download('model.pkl')
```

### Step 3: Load and Test Model

```python
text ="Hi this is input text"

#load model from notebook environment and open tokenizer
back = pickle.load(open('model.pkl', 'rb'))
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

#encode the input text
input_ids = tokenizer.encode(text, return_tensors='pt')

#pass the encoded input to the model, then decode the response
output = back.generate(input_ids, max_length = 20, num_return_sequences=1)
output = tokenizer.decode(output[0])

#print return from model
print(output)
```
